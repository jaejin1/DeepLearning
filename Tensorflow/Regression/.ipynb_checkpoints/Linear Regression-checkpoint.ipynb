```python
import tensorflow as tf

x_data = [1,2,3]
y_data = [1,2,3]

# tf.Variable: 그래프를 계산하면서 최적화 할 변수들. 신경망을 좌우하는 값들
# tf.random_normal: 각 변수들의 초기값을 정규분포 랜덤 값으로 초기화.
# tf.random_uniform: 균등분포로부터 난수값 반환
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.random_uniform([1], -1.0, 1.0))

# tf.placeholder: 계산을 실행할 때 입력값을 받는 변수로 사용.
# name: 나중에 텐서보드등으로 값의 변화 보기위해서 이름 붙여줌.
X = tf.placeholder(tf.float32, name="X")
Y = tf.placeholder(tf.float32, name="Y")
print(X)
print(Y)

hypothesis = W * X + b

# cost 함수 작성
cost = tf.reduce_mean(tf.square(hypothesis - Y))
# 경사하강법 최적화
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.1)
# 비용을 최소화 하기
train_op = optimizer.minimize(cost)

# for graph
W_val = []
cost_val2 = []

# 세션 생성하고 초기화
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    # 최적화 100번
    for step in range(100):
        # sess.run을 통해 train_op와 cost 그래프를 계산함
        # 이때 수식에 넣어야 할 값을 feed_dict을 통해 전달.
        _, cost_val = sess.run([train_op, cost], feed_dict={X: x_data, Y: y_data})
        
        print(step, cost_val, sess.run(W), sess.run(b))
        
        # for graph
        W_val.append(sess.run(W))
        cost_val2.append(cost_val)
        
    print("\n--- test ---")
    print("X: 5, Y:", sess.run(hypothesis, feed_dict={X: 5}))
    print("X: 2.5, Y:", sess.run(hypothesis, feed_dict={X: 2.5}))
```

```python
import matplotlib.pyplot as plt

plt.plot(W_val, cost_val2, 'ro')
plt.ylabel('Cost')
plt.xlabel('W')
plt.show()
```
